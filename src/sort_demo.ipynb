# Sort demo

```python
import pandas as pd
```

```python
cols = ['Vehicle Type','DirectionTime_O','GantryID_O','DirectionTime_D','GantryID_D','Trip Length','Trip End','Trip Information']
data = pd.read_csv('TDCS_M06A_20190830_080000.csv', names = cols)
```

```python
data.head()
```

```python
def my_sort(df, col, is_asc = True):
    """
    Sort function.
    Inputs:
    - df: DataFrame
            the data to sort.
    - col: str or list of str
            Sort by column(s).
    - is_asc: bool or list of bool, default True
            1/True: ascending, 0/False: descending

    Returns: DataFrame
    """
    return df.sort_values(col, ascending= is_asc)
```

```python
my_sort(data, 'Vehicle Type', 1)
```

```python
def my_search(df, col, text):
    """
    Sort function.
    Inputs:
    - df: DataFrame
            the data to search.
    - col: str or list of str
            Search by column(s).
    - text: any type
            content to search

    Returns: DataFrame
    """
    return df.loc[df[col] == text]
```

```python
my_search(data, 'Vehicle Type', 31)
```
