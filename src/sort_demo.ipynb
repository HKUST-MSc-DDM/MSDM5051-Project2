# Sort demo

```python
import pandas as pd
```

```python
cols = ['Vehicle Type','DirectionTime_O','GantryID_O','DirectionTime_D','GantryID_D','Trip Length','Trip End','Trip Information']
data = pd.read_csv('TDCS_M06A_20190830_080000.csv', names = cols)
```

```python
data.head()
```

```python
def my_sort(df, by_col, sort_type):
    """
    Sort function.
    Inputs:
    - df: DataFrame
            the data to sort.
    - by_col: str or list of str
            Sort by column(s).
    - sort_type: str or list of str
            asc: ascending, desc: descending

    Returns: DataFrame
    """
    if sort_type == 'asc':
        return df.sort_values(by_col, ascending= True)
    elif sort_type == 'desc':
        return df.sort_values(by_col, ascending= False)
    return -1
```

```python
my_sort(data, 'Vehicle Type', 'asc').head()
```

```python
def my_search(df, by_col, text):
    """
    Sort function.
    Inputs:
    - df: DataFrame
            the data to search.
    - by_col: str or list of str
            Search by column(s).
    - text: any type
            content to search

    Returns: DataFrame
    """
    return df.loc[df[by_col] == text]
```

```python
my_search(data, 'Vehicle Type', 31).head()
```
