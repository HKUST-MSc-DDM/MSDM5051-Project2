```python
from tkinter import *
from  tkinter  import ttk
import tkinter.messagebox
import tkinter.filedialog
from pandastable import Table
import pandas as pd
```

```python
def my_sort(df, by_col, sort_type):
    """
    Sort function.
    Inputs:
    - df: DataFrame
            the data to sort.
    - by_col: str or list of str
            Sort by column(s).
    - sort_type: str or list of str
            asc: ascending, desc: descending

    Returns: DataFrame
    """
    if sort_type == 'asc':
        return df.sort_values(by_col, ascending= True)
    elif sort_type == 'desc':
        return df.sort_values(by_col, ascending= False)
    return -1
```

```python
def my_search(df, by_col, text):
    """
    Sort function.
    Inputs:
    - df: DataFrame
            the data to search.
    - by_col: str or list of str
            Search by column(s).
    - text: any type
            content to search

    Returns: DataFrame
    """
    return df.loc[df[by_col] == text]
```
